```python
import pandas as pd
import plotly.express as px
from datetime import datetime
import pytz
```

```python
df = pd.read_csv("Play Store Data.csv") 
df
```

```python
now_utc = datetime.utcnow()
ist = pytz.timezone('Asia/Kolkata')
now_ist = now_utc.replace(tzinfo=pytz.utc).astimezone(ist)
if not (17 <= now_ist.hour < 22):
    print("Graph not available at this time (Only visible from 5 PM to 7 PM IST).")
else:
    # Define allowed categories
    allowed_categories = ['Game', 'Beauty', 'Business', 'Comics', 'Communication', 'Dating', 'Entertainment', 'Social', 'Events']
    
    # Apply all filters
    filtered_df = df[
        (df['Rating'] > 3.5) &
        (df['Category'].isin(allowed_categories)) &
        (df['Reviews'] > 500) &
        (~df['App'].str.contains("s", case=False)) &
        (df['Sentiment_Subjectivity'] > 0.5) &
        (df['Installs'] > 50000)
    ].copy()

    # Translate categories
    translations = {
        'Beauty': 'सौंदर्य',       # Hindi
        'Business': 'வணிகம்',     # Tamil
        'Dating': 'Partnersuche'  # German
    }
    filtered_df['Category'] = filtered_df['Category'].replace(translations)

    # Set custom colors
    colors = filtered_df['Category'].apply(lambda x: 'pink' if x == 'Game' else None)

    # Plot the bubble chart
    fig = px.scatter(
        filtered_df,
        x="Size",
        y="Rating",
        size="Installs",
        color=colors,
        hover_name="App",
        title="App Size vs Rating (Bubble size = Installs)",
        labels={"Size": "App Size (MB)", "Rating": "Average Rating"}
    )

    fig.show()
```
